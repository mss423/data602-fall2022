```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
# Get html data from url
r = requests.get('https://web.archive.org/web/20201112015618/https://www.spaceweatherlive.com/en/solar-activity/top-50-solar-flares.html')
```

```python
# Use Beautiful soup to get the data as html
root = BeautifulSoup(r.content,'html5lib')
```

```python
# Get table data and store in variable
table = root.find('table')
```

```python
# Create new dataframe with pandas with column names
new_table = pd.DataFrame(columns=['rank', 'x_class', 'date', 'region', 'start_time', 'max_time', 'end_time', 'movie'], index = range(0,50))

# Put data in table
row_marker = 0
for row in table.find_all('tr'):
    column_marker = 0
    columns = row.find_all('td')
    row_marker += 1
    for column in columns:
        if (row_marker > 1):
            new_table.iat[row_marker-2,column_marker] = column.get_text()
            column_marker += 1
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
r = requests.get('http://cdaw.gsfc.nasa.gov/CME_list/radio/waves_type2.html')
```

```python
root = BeautifulSoup(r.content,'html5lib')
```

```python
table = root.find('pre')
```

```python
row_marker = 0
content = table.get_text()
lines = content.split('\n')
for i in range(0,11):
    lines.pop(0);
lines.pop(len(lines)-1)
lines.pop(len(lines)-1)
lines.pop(len(lines)-1)
new_table = pd.DataFrame(columns=['start_date', 'start_time', 'end_date', 'end_time', 'start_frequency', 'end_frequency', 'flare_location', 'flare_region','flare_classification', 'cme_date', 'cme_time', 'cme_angle', 'cme_width', 'cme_speed'], index = range(0,len(lines)))
for line in lines:
    cols = line.split(' ')
    while '' in cols:
        cols.remove('')
    col_marker = 0
    while col_marker < 14:
        new_table.iat[row_marker, col_marker] = cols[col_marker]
        col_marker += 1
    row_marker += 1
```

```python
new_table
```

```python
# Replace all instances of '-' with '<NA>' as stated by the prompt
new_table = new_table.replace('-', '<NA>')

# Add column for Halo
# Get rows that have a Halo note
halo_idx = new_table.index[new_table['cme_angle'].str.contains('Halo')]
# print(halo_idx)
new_table['halo'] = new_table['cme_angle']
for index, row in new_table.iterrows():
  if new_table.at[index,'halo'] == 'Halo':
    new_table.at[index,'halo'] = 1
  else:
    new_table.at[index,'halo'] = 0 

new_table = new_table.replace('Halo', '<NA>')
new_table
```

```python
# First get the year info
years = new_table['start_date'].str[:4]
new_table['start_date'] = new_table['start_date'].str[5:]
# Handle 24:00 times
new_table = new_table.replace('24:00', '00:00')
# Remove missing CME time row
# new_table = new_table[new_table.cme_time != None]

# Combine the date column with the time columns to convert to datetime format
df_st = pd.to_datetime(years + '/' + new_table['start_date'] + ' ' + new_table['start_time'])
df_et = pd.to_datetime(years + '/' + new_table['end_date'] + ' ' + new_table['end_time'])

# df_ct = pd.to_datetime(years + '/' + new_table['cme_date'] + ' ' + new_table['cme_time'])

# Need to get the year info from start_date

# Add the new columns of the datetimes
new_table['start_datetime'] = df_st
new_table['end_datetime'] = df_et
# new_table['cme_datetime'] = df_ct

# Drop the old columns
new_table = new_table.drop('start_date', 1)
new_table = new_table.drop('end_date', 1)
new_table = new_table.drop('start_time', 1)
new_table = new_table.drop('end_time', 1)

# new_table = new_table.drop('cme_time', 1)
# new_table = new_table.drop('cme_date', 1)
```
